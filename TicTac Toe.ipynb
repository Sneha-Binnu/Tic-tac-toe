```python

from tkinter import *
from tkinter import messagebox
import random

# Global state
board = [""] * 9
buttons = []
current_player = "X"
mode = ""  # "pvp" or "pvc"

# Win conditions
win_cond = [
    (0, 1, 2), (3, 4, 5), (6, 7, 8),  # Rows
    (0, 3, 6), (1, 4, 7), (2, 5, 8),  # Columns
    (0, 4, 8), (2, 4, 6)              # Diagonals
]

# --- Game Logic ---
def check_winner():
    for i, j, k in win_cond:
        if board[i] == board[j] == board[k] and board[i] != "":
            return board[i]
    if "" not in board:
        return "Draw"
    return None

def show_result(winner):
    if winner == "Draw":
        messagebox.showinfo("Result", "It's a draw!")
    else:
        name = "Player 2" if mode == "pvp" else "Computer"
        winner_name = "Player 1" if winner == "X" else name
        messagebox.showinfo("Winner", f"{winner_name} wins!")
    reset_game()

def computer_move():
    empty = [i for i, val in enumerate(board) if val == ""]
    if empty:
        index = random.choice(empty)
        board[index] = "O"
        buttons[index].config(text="O", state=DISABLED)
        winner = check_winner()
        if winner:
            show_result(winner)

def button_click(index):
    global current_player
    if board[index] == "":
        board[index] = current_player
        buttons[index].config(text=current_player, state=DISABLED)
        winner = check_winner()
        if winner:
            show_result(winner)
            return
        if mode == "pvc":
            if current_player == "X":
                root.after(500, computer_move)
        else:
            current_player = "O" if current_player == "X" else "X"

def reset_game():
    global board, current_player
    board = [""] * 9
    current_player = "X"
    for btn in buttons:
        btn.config(text="", state=NORMAL)

# --- UI Management ---
def show_start_screen():
    game_frame.pack_forget()
    reset_game()
    start_frame.pack()

def start_game(selected_mode):
    global mode
    mode = selected_mode

    # Show instructions
    if mode == "pvp":
        msg = (
            "🎮 Player vs Player Mode\n\n"
            "• Player 1 is X\n"
            "• Player 2 is O\n"
            "• Take turns to tap a box\n"
            "• First to align 3 marks wins!"
        )
    else:
        msg = (
            "🤖 Player vs Computer Mode\n\n"
            "• You are X\n"
            "• Computer is O\n"
            "• Tap to play your turn\n"
            "• Computer will auto-play its turn"
        )

    messagebox.showinfo("Game Instructions", msg)

    # Show game screen
    start_frame.pack_forget()
    game_frame.pack()

def setup_main_window():
    global root, start_frame, game_frame

    root = Tk()
    root.title("Tic Tac Toe")
    root.geometry("320x400")
    root.resizable(False, False)

    # --- Start Screen ---
    start_frame = Frame(root)
    Label(start_frame, text="Tic Tac Toe", font=("Arial", 18, "bold")).pack(pady=20)
    Label(start_frame, text="Choose Mode:", font=("Arial", 14)).pack(pady=10)

    Button(start_frame, text="Player vs Player", font=("Arial", 12),
           width=20, command=lambda: start_game("pvp")).pack(pady=5)

    Button(start_frame, text="Player vs Computer", font=("Arial", 12),
           width=20, command=lambda: start_game("pvc")).pack(pady=5)

    start_frame.pack()

    # --- Game Screen ---
    game_frame = Frame(root)

    # Game board buttons
    for i in range(9):
        btn = Button(game_frame, text="", font=("Arial", 20), width=5, height=2,
                     command=lambda i=i: button_click(i))
        btn.grid(row=i // 3, column=i % 3, padx=2, pady=2)
        buttons.append(btn)

    # Control buttons (Reset and Back)
    Button(game_frame, text="Reset Game", font=("Arial", 12),
           command=reset_game).grid(row=3, column=0, columnspan=3, pady=10)

    Button(game_frame, text="Back to Menu", font=("Arial", 12),
           command=show_start_screen).grid(row=4, column=0, columnspan=3)

    root.mainloop()

# Start app
setup_main_window()
```
